# SVM (Support Vector Machine)
It's a supervised learning algorithm for classification problems that classifies cases by finding a suitable separator (_Hyper Plane_)

Common terms:
- margin
- hyperplane
- support vector



Margine can be soft and hard.
- soft margin: that allows misclassification


>How do we select the best soft margin?

Using cross validation.

### Kernel

It's very helpful to transfer data from lower dimension to higher dimension

--O--O-O-O-----X-X--X---O--O-O--
If we have data like this, One type of data is in the middle and other type of data is on the both side, we can't use SVC (Support Vector Classifier) here. We need to transform the data in the higher dimension. That is what **kernel** do.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import seaborn as sns
style.use("default")
sns.set()
```

```python
df = pd.read_csv("recipes_muffins_cupcakes.csv")
df.head()
```

```python
sns.lmplot(x="Sugar", y="Butter", data=df);
```

```python
sns.lmplot(x="Sugar", y="Butter", data=df, hue="Type");
```

```python
sns.lmplot(x="Sugar", y="Butter", data=df, hue="Type", fit_reg=False);
```

```python
sns.lmplot(x="Sugar", y="Butter", data=df, hue="Type", fit_reg=False, scatter_kws={"s": 70});
```
