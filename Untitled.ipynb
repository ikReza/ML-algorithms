# Linear Regression

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("accidents.csv", index_col=0)
data
```

```python
X = data["Number of Accidents"].values
y = data["Death"].values
```

```python
from matplotlib import style

style.use("default")
plt.scatter(X, y, label="scatter plot", c="blue")
plt.plot(np.unique(X), np.poly1d(np.polyfit(X, y, 1))(np.unique(X)), c="red", label="best fit line")

plt.xlabel("Number of accidents")
plt.ylabel("Death")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = X.reshape(-1, 1)
print(X)
reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
print(y_pred)
reg.score(X, y)*100
```

```python
plt.scatter(X, y_pred)
plt.show()
```
