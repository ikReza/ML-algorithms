# Activation function

## `Sigmoid` function



It is considered bad for deep neural network. Because it's mean value is `0.5`. But we want out data to be 0 centered.

The sigmoid function is not symmetric around zero. So output of all the neurons will be of the same sign.

## `tanh` function



This is the solution of the above mentioned problem with `sigmoid` function

**Problem:** Though `tanh` is slightly better than `sigmoid`, both have similar problem called **`vanishing gradient problem`**

#### vanishing gradient

- output = $ \sigma(...\sigma(...\sigma(...\sigma(...\sigma(...))))) $
- we end up multiplying the derivative of the sigmoid over and over again

> what's wrong with the derivative of the sigmoid?

**Ans:** Derivative of sigmoid is very tiny number(maximum=0.25). So, if we multiply this small number over and over again, we'll get a number close to 0.
- $ 0.25^5 \approx 0.001 $
- in reality, this value will not be always 0.25, something less than 0.25. e.g. $ 0.1^5 \approx 0.00001 $
- we know, the property of **gradient descent** is that it takes baby steps when derivative close to 0. This means model will be trained slowly. 


So, don't use activation function that has _vanishing gradient_ property
