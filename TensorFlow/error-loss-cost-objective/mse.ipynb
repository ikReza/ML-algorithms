# MSE (Mean Squared Error)

$ mse = \frac{1}{N} \sum \limits_{i=1}^{N} (y_{i} - \hat y_{i})^2 $

### Q-1: Why is it squared?

- We want the error to be positive
- If errors are both positve and negative, they will cancel out each other

```python
y_true = [5, 8, 12, 6]
y_pred = [3, 13, 14, 1]
sum_error = 0
for i in range(len(y_true)):
    err = y_true[i] - y_pred[i]
    sum_error += err
    print(f"Error: {err}")
print(f"Total Error: {sum_error}")
```

### Q-1: Why not MAE (mean absolute error)?

$ mae = \frac{1}{N} \sum \limits_{i=1}^{N} |(y_{i} - \hat y_{i})| $

```python
for i in range(len(y_true)):
    err = abs(y_true[i] - y_pred[i])
    sum_error += err
    print(f"Error: {err}")
print(f"Total Error: {sum_error}, MAE: {sum_error/len(y_true)}")
```

- mse is preferred in regression problem

### Some theory

There are three metrics which are generally used for evaluation of Regression problems (like Linear Regression, Decision Tree Regression, Random Forest Regression etc.):
- MAE
- MSE
- RMSE
