# Linear Regression

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("accidents.csv", index_col=0)
data
```

```python
X = data["Number of Accidents"].values
y = data["Death"].values
```

```python
from matplotlib import style

style.use("default")
plt.scatter(X, y, label="scatter plot", c="blue")
#best fit line
plt.plot(np.unique(X), np.poly1d(np.polyfit(X, y, 1))(np.unique(X)), c="red", label="best fit line")

plt.xlabel("Number of accidents")
plt.ylabel("Death")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = X.reshape(-1, 1)
reg = LinearRegression()
reg.fit(X, y)
#predict takes 2D array
reg.predict([[3381]])
```

```python
reg.score(X, y)*100
```

## Teaching machine to calculate sum of 2 numbers

```python
df2 = pd.read_csv("sum.csv")
df2
```

```python
features = df2[["a", "b"]]  #X
labels = df2["sum"]  #y
```

```python
reg2 = LinearRegression()
reg2.fit(features, labels)
```

```python
reg2.predict([[10, 14]])
```

```python
reg2.score(features, labels)
```
